```python
# -*- coding: utf-8 -*-

# #--- ì´ˆê¸° íŒ¨í‚¤ì§€ ì„¤ì¹˜ ---#
# !python -m pip install --upgrade pip
# !pip install pandas selenium==4.18.1 webdriver-manager undetected-chromedriver user_agents pyperclip chromedriver_autoinstaller

#--- ë¼ì´ë¸ŒëŸ¬ë¦¬ import ---#
import pandas as pd
import warnings
import os
import time
import shutil
import urllib
import random

warnings.filterwarnings(action='ignore')

from PIL import Image
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, WebDriverException

import chromedriver_autoinstaller
import undetected_chromedriver as uc
from user_agents import parse

# ğŸš€ ê°œì„ ì‚¬í•­ 1: ì„¤ì •ê°’ ë¶„ë¦¬ - ëª¨ë“  ì„¤ì •ì„ í•œ ê³³ì—ì„œ ê´€ë¦¬
CONFIG = {
    "WAIT_TIMEOUT": 10,
    "RETRY_COUNT": 3,
    "MIN_DELAY": 3,
    "MAX_DELAY": 8,
    "POPUP_WAIT": 5,
    "USER_AGENT": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
    "SAVE_IMAGES": True,
    "SAVE_INTERMEDIATE": True,
    "MAX_PRODUCT_NAME_LENGTH": 30
}

# ğŸ¯ ê²€ìƒ‰í•  ë„ì‹œë“¤ ì„¤ì •
CITIES_TO_SEARCH = ["ë°©ì½•", "íŒŒë¦¬"]

# ğŸ†• ë„ì‹œë³„ ìˆ˜ì§‘í•  ìƒí’ˆ ê°œìˆ˜ ì„¤ì •
PRODUCTS_PER_CITY = 3

# ğŸŒ ë„ì‹œë³„ ëŒ€ë¥™ ë° êµ­ê°€ ì •ë³´ ë§¤í•‘
CITY_INFO = {
    # ğŸŒ ë„ì‹œë³„ ëŒ€ë¥™ ë° êµ­ê°€ ì •ë³´ ë§¤í•‘ (ì „ì²´ ë²„ì „)
CITY_INFO = {
    # ì•„ì‹œì•„
    "ë°©ì½•": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "ë„ì¿„": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì¼ë³¸"},
    "ì˜¤ì‚¬ì¹´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì¼ë³¸"},
    "êµí† ": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì¼ë³¸"},
    "ì‹±ê°€í¬ë¥´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì‹±ê°€í¬ë¥´"},
    "í™ì½©": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "í™ì½©"},
    "íƒ€ì´ë² ì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ëŒ€ë§Œ"},
    "ìƒí•˜ì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì¤‘êµ­"},
    "ë² ì´ì§•": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì¤‘êµ­"},
    "í‘¸ì¼“": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "íŒŒíƒ€ì•¼": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "ì¹˜ì•™ë§ˆì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "í˜¸ì¹˜ë¯¼": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ë² íŠ¸ë‚¨"},
    "í•˜ë…¸ì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ë² íŠ¸ë‚¨"},
    "ë‹¤ë‚­": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ë² íŠ¸ë‚¨"},
    "ì„¸ë¶€": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "í•„ë¦¬í•€"},
    "ë³´ë¼ì¹´ì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "í•„ë¦¬í•€"},
    "ë°œë¦¬": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì¸ë„ë„¤ì‹œì•„"},
    "ìì¹´ë¥´íƒ€": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ì¸ë„ë„¤ì‹œì•„"},
    "ì¿ ì•Œë¼ë£¸í‘¸ë¥´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ë§ë ˆì´ì‹œì•„"},
    "ì½”íƒ€í‚¤ë‚˜ë°œë£¨": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ë§ë ˆì´ì‹œì•„"},
    "ë‘ì¹´ìœ„": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ë§ë ˆì´ì‹œì•„"},
    "í˜ë‚­": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ë§ë ˆì´ì‹œì•„"},
    "ì•„ìœ íƒ€ì•¼": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "ìˆ˜ì½”íƒ€ì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "ì¹´ì˜¤ì•¼ì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "í›„ì•„íŒ": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "ì½”ì‚¬ë¬´ì´": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "í¬ë¼ë¹„": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "íƒœêµ­"},
    "ëª°ë””ë¸Œ": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "ëª°ë””ë¸Œ"},
    "ì œì£¼ë„": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "í•œêµ­"},
    "ë¶€ì‚°": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "í•œêµ­"},
    "ê°•ë¦‰": {"ëŒ€ë¥™": "ì•„ì‹œì•„", "êµ­ê°€": "í•œêµ­"},
    
    # ìœ ëŸ½
    "íŒŒë¦¬": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "í”„ë‘ìŠ¤"},
    "ëŸ°ë˜": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ì˜êµ­"},
    "ë¡œë§ˆ": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ì´íƒˆë¦¬ì•„"},
    "ë°€ë¼ë…¸": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ì´íƒˆë¦¬ì•„"},
    "ë² ë‹ˆìŠ¤": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ì´íƒˆë¦¬ì•„"},
    "ë°”ë¥´ì…€ë¡œë‚˜": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ìŠ¤í˜ì¸"},
    "ë§ˆë“œë¦¬ë“œ": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ìŠ¤í˜ì¸"},
    "ì•”ìŠ¤í…Œë¥´ë‹´": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ë„¤ëœë€ë“œ"},
    "ë² ë¥¼ë¦°": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ë…ì¼"},
    "ë®Œí—¨": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ë…ì¼"},
    "í”„ë¼í•˜": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ì²´ì½”"},
    "ë¹„ì—”ë‚˜": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ì˜¤ìŠ¤íŠ¸ë¦¬ì•„"},
    "ì·¨ë¦¬íˆ": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ìŠ¤ìœ„ìŠ¤"},
    "ì œë„¤ë°”": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ìŠ¤ìœ„ìŠ¤"},
    "ìŠ¤í†¡í™€ë¦„": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ìŠ¤ì›¨ë´"},
    "ì½”íœí•˜ê²": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ë´ë§ˆí¬"},
    "í—¬ì‹±í‚¤": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "í•€ë€ë“œ"},
    "ëª¨ìŠ¤í¬ë°”": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ëŸ¬ì‹œì•„"},
    "ìƒíŠ¸í˜í…Œë¥´ë¶€ë¥´í¬": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ëŸ¬ì‹œì•„"},
    "ì•„í…Œë„¤": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "ê·¸ë¦¬ìŠ¤"},
    "ë¦¬ìŠ¤ë³¸": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "í¬ë¥´íˆ¬ê°ˆ"},
    "ë¶€ë‹¤í˜ìŠ¤íŠ¸": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "í—ê°€ë¦¬"},
    "ë°”ë¥´ìƒ¤ë°”": {"ëŒ€ë¥™": "ìœ ëŸ½", "êµ­ê°€": "í´ë€ë“œ"},
    
    # ë¶ë¯¸
    "ë‰´ìš•": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ë¡œìŠ¤ì•¤ì ¤ë ˆìŠ¤": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ë¼ìŠ¤ë² ì´ê±°ìŠ¤": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ìƒŒí”„ë€ì‹œìŠ¤ì½”": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ì‹œì¹´ê³ ": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ë³´ìŠ¤í„´": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ë§ˆì´ì• ë¯¸": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ì‹œì• í‹€": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "í•˜ì™€ì´": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë¯¸êµ­"},
    "ë°´ì¿ ë²„": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ìºë‚˜ë‹¤"},
    "í† ë¡ í† ": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ìºë‚˜ë‹¤"},
    "ëª¬íŠ¸ë¦¬ì˜¬": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ìºë‚˜ë‹¤"},
    "ë©•ì‹œì½”ì‹œí‹°": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë©•ì‹œì½”"},
    "ì¹¸ì¿¤": {"ëŒ€ë¥™": "ë¶ë¯¸", "êµ­ê°€": "ë©•ì‹œì½”"},
    
    # ë‚¨ë¯¸
    "ë¦¬ìš°ë°ìë„¤ì´ë£¨": {"ëŒ€ë¥™": "ë‚¨ë¯¸", "êµ­ê°€": "ë¸Œë¼ì§ˆ"},
    "ìƒíŒŒìš¸ë£¨": {"ëŒ€ë¥™": "ë‚¨ë¯¸", "êµ­ê°€": "ë¸Œë¼ì§ˆ"},
    "ë¶€ì—ë…¸ìŠ¤ì•„ì´ë ˆìŠ¤": {"ëŒ€ë¥™": "ë‚¨ë¯¸", "êµ­ê°€": "ì•„ë¥´í—¨í‹°ë‚˜"},
    "ë¦¬ë§ˆ": {"ëŒ€ë¥™": "ë‚¨ë¯¸", "êµ­ê°€": "í˜ë£¨"},
    "ì‚°í‹°ì•„ê³ ": {"ëŒ€ë¥™": "ë‚¨ë¯¸", "êµ­ê°€": "ì¹ ë ˆ"},
    "ë³´ê³ íƒ€": {"ëŒ€ë¥™": "ë‚¨ë¯¸", "êµ­ê°€": "ì½œë¡¬ë¹„ì•„"},
    "í‚¤í† ": {"ëŒ€ë¥™": "ë‚¨ë¯¸", "êµ­ê°€": "ì—ì½°ë„ë¥´"},
    
    # ì˜¤ì„¸ì•„ë‹ˆì•„
    "ì‹œë“œë‹ˆ": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "í˜¸ì£¼"},
    "ë©œë²„ë¥¸": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "í˜¸ì£¼"},
    "ê³¨ë“œì½”ìŠ¤íŠ¸": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "í˜¸ì£¼"},
    "í¼ìŠ¤": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "í˜¸ì£¼"},
    "ì• ë“¤ë ˆì´ë“œ": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "í˜¸ì£¼"},
    "ì¼€ì–¸ì¦ˆ": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "í˜¸ì£¼"},
    "ì˜¤í´ëœë“œ": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "ë‰´ì§ˆëœë“œ"},
    "í¬ë¼ì´ìŠ¤íŠ¸ì²˜ì¹˜": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "ë‰´ì§ˆëœë“œ"},
    "ì›°ë§í„´": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "ë‰´ì§ˆëœë“œ"},
    "ê´Œ": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "ê´Œ"},
    "ì‚¬ì´íŒ": {"ëŒ€ë¥™": "ì˜¤ì„¸ì•„ë‹ˆì•„", "êµ­ê°€": "ì‚¬ì´íŒ"},
    
    # ì¤‘ë™
    "ë‘ë°”ì´": {"ëŒ€ë¥™": "ì¤‘ë™", "êµ­ê°€": "UAE"},
    "ì•„ë¶€ë‹¤ë¹„": {"ëŒ€ë¥™": "ì¤‘ë™", "êµ­ê°€": "UAE"},
    "ë„í•˜": {"ëŒ€ë¥™": "ì¤‘ë™", "êµ­ê°€": "ì¹´íƒ€ë¥´"},
    "ì´ìŠ¤íƒ„ë¶ˆ": {"ëŒ€ë¥™": "ì¤‘ë™", "êµ­ê°€": "í„°í‚¤"},
    "í…”ì•„ë¹„ë¸Œ": {"ëŒ€ë¥™": "ì¤‘ë™", "êµ­ê°€": "ì´ìŠ¤ë¼ì—˜"},
    "ì˜ˆë£¨ì‚´ë ˜": {"ëŒ€ë¥™": "ì¤‘ë™", "êµ­ê°€": "ì´ìŠ¤ë¼ì—˜"},
    
    # ì•„í”„ë¦¬ì¹´
    "ì¹´ì´ë¡œ": {"ëŒ€ë¥™": "ì•„í”„ë¦¬ì¹´", "êµ­ê°€": "ì´ì§‘íŠ¸"},
    "ì¼€ì´í”„íƒ€ìš´": {"ëŒ€ë¥™": "ì•„í”„ë¦¬ì¹´", "êµ­ê°€": "ë‚¨ì•„í”„ë¦¬ì¹´ê³µí™”êµ­"},
    "ìš”í•˜ë„¤ìŠ¤ë²„ê·¸": {"ëŒ€ë¥™": "ì•„í”„ë¦¬ì¹´", "êµ­ê°€": "ë‚¨ì•„í”„ë¦¬ì¹´ê³µí™”êµ­"},
    "ì¹´ì‚¬ë¸”ë‘ì¹´": {"ëŒ€ë¥™": "ì•„í”„ë¦¬ì¹´", "êµ­ê°€": "ëª¨ë¡œì½”"},
    "ë§ˆë¼ì¼€ì‹œ": {"ëŒ€ë¥™": "ì•„í”„ë¦¬ì¹´", "êµ­ê°€": "ëª¨ë¡œì½”"},
    "ë‚˜ì´ë¡œë¹„": {"ëŒ€ë¥™": "ì•„í”„ë¦¬ì¹´", "êµ­ê°€": "ì¼€ëƒ"},
    "ë‹¤ë¥´ì—ìŠ¤ì‚´ëŒ": {"ëŒ€ë¥™": "ì•„í”„ë¦¬ì¹´", "êµ­ê°€": "íƒ„ìë‹ˆì•„"},
}

# --- í•¨ìˆ˜ ì •ì˜ ---

def print_progress(current, total, city_name, status="ì§„í–‰ì¤‘"):
    percentage = (current / total) * 100
    bar_length = 30
    filled_length = int(bar_length * current // total)
    bar = 'â–ˆ' * filled_length + 'â–‘' * (bar_length - filled_length)
    emoji = "ğŸ”" if status == "ì§„í–‰ì¤‘" else "âœ…" if status == "ì™„ë£Œ" else "âŒ"
    print(f"\n{emoji} ì§„í–‰ë¥ : [{bar}] {percentage:.1f}% ({current}/{total})")
    print(f"ğŸ“ í˜„ì¬ ì‘ì—…: {city_name} - {status}")
    print("-" * 50)

def save_intermediate_results(results, city_name):
    if results and CONFIG["SAVE_INTERMEDIATE"]:
        try:
            timestamp = time.strftime('%Y%m%d_%H%M%S')
            temp_filename = f"temp_ì¤‘ê°„ì €ì¥_{city_name}_{timestamp}.csv"
            pd.DataFrame(results).to_csv(temp_filename, index=False, encoding='utf-8-sig')
            print(f"  ğŸ’¾ ì¤‘ê°„ ê²°ê³¼ ì €ì¥: {temp_filename}")
        except Exception as e:
            print(f"  âš ï¸ ì¤‘ê°„ ì €ì¥ ì‹¤íŒ¨: {e}")

def retry_operation(func, operation_name, max_retries=None):
    if max_retries is None:
        max_retries = CONFIG["RETRY_COUNT"]
    for attempt in range(max_retries):
        try:
            return func()
        except (TimeoutException, NoSuchElementException, WebDriverException) as e:
            if attempt == max_retries - 1:
                print(f"  âŒ {operation_name} ìµœì¢… ì‹¤íŒ¨: {type(e).__name__}")
                raise e
            print(f"  ğŸ”„ {operation_name} ì¬ì‹œë„ {attempt + 1}/{max_retries} (ì˜¤ë¥˜: {type(e).__name__})")
            time.sleep(2)
        except Exception as e:
            print(f"  âŒ {operation_name} ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜: {type(e).__name__}: {e}")
            raise e

def get_product_name(driver, config):
    title_selectors = [
        (By.CSS_SELECTOR, "h1"),
        (By.XPATH, "/html/body/div[1]/main/div[1]/section/div[1]/h1")
    ]
    for selector_type, selector_value in title_selectors:
        try:
            return WebDriverWait(driver, config["WAIT_TIMEOUT"]).until(EC.presence_of_element_located((selector_type, selector_value))).text
        except TimeoutException:
            continue
    raise NoSuchElementException("ìƒí’ˆëª…ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")

def get_price_info(driver, config):
    price_selectors = [
        (By.CSS_SELECTOR, "span.price"),
        (By.XPATH, "//span[contains(text(), 'ì›')]")
    ]
    for selector_type, selector_value in price_selectors:
        try:
            return WebDriverWait(driver, config["WAIT_TIMEOUT"]).until(EC.presence_of_element_located((selector_type, selector_value))).text
        except TimeoutException:
            continue
    return "ì •ë³´ ì—†ìŒ"

def get_rating_info(driver, config):
    rating_selectors = [
        (By.CSS_SELECTOR, "span.rating"),
        (By.XPATH, "//span[contains(@class, 'rating')]")
    ]
    for selector_type, selector_value in rating_selectors:
        try:
            return WebDriverWait(driver, config["WAIT_TIMEOUT"]).until(EC.presence_of_element_located((selector_type, selector_value))).text
        except TimeoutException:
            continue
    return "ì •ë³´ ì—†ìŒ"

def get_review_count_info(driver):
    review_count_selectors = [
        (By.XPATH, "//span[contains(text(), '(') and contains(text(), ')')]"),
        (By.CSS_SELECTOR, "span[class*='rating'] + span")
    ]
    for selector_type, selector_value in review_count_selectors:
        try:
            return WebDriverWait(driver, 3).until(EC.presence_of_element_located((selector_type, selector_value))).text
        except TimeoutException:
            continue
    return ""

def get_language_info(driver):
    language_keywords = ['í•œêµ­ì–´', 'ì˜ì–´', 'Korean', 'English']
    for keyword in language_keywords:
        try:
            return WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, f"//span[contains(text(), '{keyword}')]"))).text
        except TimeoutException:
            continue
    return ""

def download_product_image(driver, config, city_name, product_name):
    if not config["SAVE_IMAGES"]:
        return "ì´ë¯¸ì§€ ì €ì¥ ë¹„í™œì„±í™”"
    try:
        img_element = WebDriverWait(driver, config["WAIT_TIMEOUT"]).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div[class*='Gallery'] img")))
        img_url = img_element.get_attribute('src')
        if img_url:
            safe_product_name = make_safe_filename(f"{city_name}_{product_name[:config['MAX_PRODUCT_NAME_LENGTH']]}")
            img_download_path = os.path.join(os.getcwd(), "myrealtrip_images", f"{safe_product_name}.png")
            urllib.request.urlretrieve(img_url, img_download_path)
            return img_download_path
    except Exception as e:
        print(f"  âš ï¸ ì´ë¯¸ì§€ ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨: {e}")
    return "ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨"

def make_safe_filename(filename):
    return "".join([c for c in filename if c.isalnum() or c in (' ', '_')]).rstrip()

def get_city_info(city_name):
    return CITY_INFO.get(city_name, {"ëŒ€ë¥™": "ê¸°íƒ€", "êµ­ê°€": "ê¸°íƒ€"}).values()

def click_product_by_index(driver, config, product_index):
    print(f"  ğŸ¯ {product_index + 1}ë²ˆì§¸ ìƒí’ˆ ì°¾ëŠ” ì¤‘...")
    # ë” ì•ˆì •ì ì¸ XPathë¡œ ìƒí’ˆ ëª©ë¡ ì „ì²´ë¥¼ ë¨¼ì € ì°¾ê³ , ê·¸ ì•ˆì—ì„œ ì¸ë±ìŠ¤ë¡œ ì ‘ê·¼
    product_list_xpath = "//div[contains(@class, 'container-xxl')]//a[contains(@href, '/offers/')]"
    try:
        products = WebDriverWait(driver, config["WAIT_TIMEOUT"]).until(
            EC.presence_of_all_elements_located((By.XPATH, product_list_xpath))
        )
        if product_index < len(products):
            product_element = products[product_index]
            driver.execute_script("arguments[0].scrollIntoView(true);", product_element)
            time.sleep(1)
            product_element.click()
            print(f"  âœ… {product_index + 1}ë²ˆì§¸ ìƒí’ˆ í´ë¦­ ì„±ê³µ!")
            time.sleep(random.uniform(config["MIN_DELAY"], config["MAX_DELAY"]))
            return True
        else:
            raise NoSuchElementException(f"ìƒí’ˆ ëª©ë¡ì— {product_index + 1}ë²ˆì§¸ ìƒí’ˆì´ ì—†ìŠµë‹ˆë‹¤. (ì´ {len(products)}ê°œ)")
    except (TimeoutException, Exception) as e:
        print(f"  âŒ ìƒí’ˆ í´ë¦­ ì‹¤íŒ¨: {type(e).__name__}")
        raise e

def go_back_to_product_list(driver):
    print(f"  â†©ï¸ ìƒí’ˆ ëª©ë¡ìœ¼ë¡œ ëŒì•„ê°€ëŠ” ì¤‘...")
    try:
        driver.back()
        time.sleep(random.uniform(3, 5))
        # í˜ì´ì§€ ë¡œë”© í™•ì¸
        WebDriverWait(driver, CONFIG["WAIT_TIMEOUT"]).until(EC.presence_of_element_located((By.CSS_SELECTOR, "input[placeholder*='ì–´ë””ë¡œ']")))
        print(f"  âœ… ìƒí’ˆ ëª©ë¡ìœ¼ë¡œ ëŒì•„ê°€ê¸° ì„±ê³µ!")
    except Exception as e:
        print(f"  âš ï¸ ë’¤ë¡œê°€ê¸° ì‹¤íŒ¨, ê²€ìƒ‰ í˜ì´ì§€ë¡œ ê°•ì œ ì´ë™: {type(e).__name__}")
        driver.get("https://www.myrealtrip.com/experiences/")
        time.sleep(random.uniform(3, 5))

# --- í¬ë¡¬ ë“œë¼ì´ë²„ ì„¤ì • ë° ì‹¤í–‰ ---
all_results = []
chromedriver_autoinstaller.install()
options = uc.ChromeOptions()
options.add_argument(f"--user-agent={CONFIG['USER_AGENT']}")
options.add_argument("--start-maximized")

# ì´ë¯¸ì§€ í´ë” ìƒì„±
if CONFIG["SAVE_IMAGES"]:
    os.makedirs(os.path.join(os.getcwd(), "myrealtrip_images"), exist_ok=True)

try:
    driver = uc.Chrome(options=options)
    print("âœ… í¬ë¡¬ ë“œë¼ì´ë²„ ì‹¤í–‰ ì„±ê³µ!")
except Exception as e:
    print(f"\nâŒ í¬ë¡¬ ë“œë¼ì´ë²„ ì‹¤í–‰ ì‹¤íŒ¨: {e}")
    print("ë¸”ë¡œê·¸(https://appfollow.tistory.com/102)ë¥¼ ì°¸ê³ í•˜ê±°ë‚˜, í¬ë¡¬ ë¸Œë¼ìš°ì €ë¥¼ ìµœì‹  ë²„ì „ìœ¼ë¡œ ì—…ë°ì´íŠ¸ í•´ë³´ì„¸ìš”.")
    exit()

# --- ë©”ì¸ í¬ë¡¤ë§ ë¡œì§ ---
print(f"ğŸŒ ì´ {len(CITIES_TO_SEARCH)}ê°œ ë„ì‹œ ê²€ìƒ‰ ì‹œì‘: {', '.join(CITIES_TO_SEARCH)}")

for city_index, city_name in enumerate(CITIES_TO_SEARCH):
    print_progress(city_index, len(CITIES_TO_SEARCH), city_name, "ì§„í–‰ì¤‘")
    continent, country = get_city_info(city_name)
    
    try:
        # 1. ë©”ì¸ ê²€ìƒ‰ í˜ì´ì§€ ì´ë™
        retry_operation(lambda: driver.get("https://www.myrealtrip.com/experiences/"), "ë©”ì¸ í˜ì´ì§€ ì´ë™")
        
        # 2. ë„ì‹œ ê²€ìƒ‰
        def search_city():
            search_input = WebDriverWait(driver, CONFIG["WAIT_TIMEOUT"]).until(EC.presence_of_element_located((By.CSS_SELECTOR, "input[placeholder*='ì–´ë””ë¡œ']")))
            search_input.clear()
            search_input.send_keys(city_name)
            time.sleep(1)
            search_input.send_keys(Keys.ENTER)
            # ê²€ìƒ‰ ê²°ê³¼ í˜ì´ì§€ ë¡œë”© í™•ì¸
            WebDriverWait(driver, CONFIG["WAIT_TIMEOUT"]).until(EC.presence_of_element_located((By.XPATH, f"//h1[contains(text(),'{city_name}')]")))
            return True
        retry_operation(search_city, f"'{city_name}' ê²€ìƒ‰")

        # 3. ì—¬ëŸ¬ ìƒí’ˆ ìˆ˜ì§‘
        print(f"  ğŸ“‹ {PRODUCTS_PER_CITY}ê°œ ìƒí’ˆ ìˆ˜ì§‘ ì‹œì‘...")
        for product_index in range(PRODUCTS_PER_CITY):
            try:
                print(f"\n  ğŸ”„ === {product_index + 1}ë²ˆì§¸ ìƒí’ˆ ì²˜ë¦¬ ì‹œì‘ ===")
                
                # ìƒí’ˆ í´ë¦­
                retry_operation(lambda: click_product_by_index(driver, CONFIG, product_index), f"{product_index + 1}ë²ˆì§¸ ìƒí’ˆ í´ë¦­")

                product_name = get_product_name(driver, CONFIG)
                price = get_price_info(driver, CONFIG)
                grade_review = get_rating_info(driver, CONFIG)
                review_count = get_review_count_info(driver)
                language = get_language_info(driver)
                img_path = download_product_image(driver, CONFIG, city_name, product_name)
                current_url = driver.current_url

                result = {
                    'ë²ˆí˜¸': len(all_results) + 1, 'ëŒ€ë¥™': continent, 'êµ­ê°€': country, 'ë„ì‹œ': city_name,
                    'ìƒí’ˆìˆœì„œ': product_index + 1, 'ìƒí’ˆëª…': product_name, 'ê°€ê²©': price, 'í‰ì ': grade_review,
                    'ë¦¬ë·°ìˆ˜': review_count, 'ì–¸ì–´': language, 'ì´ë¯¸ì§€_ê²½ë¡œ': img_path, 'URL': current_url,
                    'ìˆ˜ì§‘_ì‹œê°„': time.strftime('%Y-%m-%d %H:%M:%S'), 'ìƒíƒœ': 'ì™„ì „ìˆ˜ì§‘'
                }
                all_results.append(result)
                print(f"  âœ… '{product_name[:30]}...' ì •ë³´ ìˆ˜ì§‘ ì™„ë£Œ")

                # ëª©ë¡ìœ¼ë¡œ ëŒì•„ê°€ê¸° (ë§ˆì§€ë§‰ ìƒí’ˆì´ ì•„ë‹ ê²½ìš°)
                if product_index < PRODUCTS_PER_CITY - 1:
                    go_back_to_product_list(driver)

            except Exception as e:
                print(f"  âŒ {product_index + 1}ë²ˆì§¸ ìƒí’ˆ ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
                if product_index < PRODUCTS_PER_CITY - 1:
                    try:
                        go_back_to_product_list(driver)
                    except Exception as back_e:
                        print(f"  âš ï¸ ëª©ë¡ìœ¼ë¡œ ë³µê·€ ì‹¤íŒ¨, ë‹¤ìŒ ë„ì‹œë¡œ ë„˜ì–´ê°‘ë‹ˆë‹¤: {back_e}")
                        break # í˜„ì¬ ë„ì‹œì˜ ë£¨í”„ë¥¼ ì¤‘ë‹¨
                continue # ë‹¤ìŒ ìƒí’ˆìœ¼ë¡œ ì§„í–‰
        
        save_intermediate_results(all_results, f"{city_name}_ì™„ë£Œ")
        print_progress(city_index + 1, len(CITIES_TO_SEARCH), city_name, "ì™„ë£Œ")

    except Exception as e:
        print(f"âŒ '{city_name}' ë„ì‹œ ì „ì²´ ì²˜ë¦¬ ì‹¤íŒ¨: {e}")
        print_progress(city_index + 1, len(CITIES_TO_SEARCH), city_name, "ì‹¤íŒ¨")
        continue

# --- ìµœì¢… ê²°ê³¼ ì €ì¥ ë° ì¢…ë£Œ ---
print("\n" + "="*60 + "\nğŸ‰ ëª¨ë“  ë„ì‹œ ê²€ìƒ‰ ì™„ë£Œ! ğŸ‰\n" + "="*60)

if all_results:
    df = pd.DataFrame(all_results)
    timestamp = time.strftime('%Y%m%d_%H%M%S')
    filename = f"myrealtrip_ìµœì¢…ê²°ê³¼_{timestamp}.csv"
    df.to_csv(filename, index=False, encoding='utf-8-sig')
    print(f"ğŸ’¾ ìµœì¢… ê²°ê³¼ {len(all_results)}ê°œê°€ '{filename}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤!")
    print(df.head().to_string())
else:
    print("âŒ ìˆ˜ì§‘ëœ ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤.")

driver.quit()
print("\nğŸšª ì›¹ë“œë¼ì´ë²„ë¥¼ ì¢…ë£Œí–ˆìŠµë‹ˆë‹¤. ëª¨ë“  ì‘ì—… ì™„ë£Œ!")
```
