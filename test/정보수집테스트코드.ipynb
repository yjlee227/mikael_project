```python
import requests
import xml.etree.ElementTree as ET

def check_osaka_coverage():
    print("ğŸ” ì˜¤ì‚¬ì¹´ KLOOK ìƒí’ˆ ì»¤ë²„ë¦¬ì§€ ë¶„ì„...")

    # sitemapì—ì„œ ì˜¤ì‚¬ì¹´ ê´€ë ¨ URL ìˆ˜ì§‘
    response = requests.get("https://www.klook.com/ko/sitemap-experiences-activity-plain.xml")
    root = ET.fromstring(response.content)

    osaka_keywords = ['osaka', 'ì˜¤ì‚¬ì¹´', 'japan', 'ì¼ë³¸']
    osaka_urls = []

    for url_elem in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
        url = url_elem.text
        if any(keyword in url.lower() for keyword in osaka_keywords):
            osaka_urls.append(url)

    print(f"ğŸ“Š ì˜¤ì‚¬ì¹´ ê´€ë ¨ URL: {len(osaka_urls)}ê°œ ë°œê²¬")

    # URL íŒ¨í„´ ë¶„ì„
    url_patterns = {}
    for url in osaka_urls:
        if '/activity/' in url:
            activity_id = url.split('activity/')[-1].split('-')[0]
            url_patterns[activity_id] = url

    print(f"ğŸ“‹ ê³ ìœ  ì•¡í‹°ë¹„í‹°: {len(url_patterns)}ê°œ")
    print("ğŸ”— ìƒ˜í”Œ URL:")
    for i, url in enumerate(list(osaka_urls)[:5]):
        print(f"  {i+1}. {url}")

    return osaka_urls

# ì‹¤í–‰í•´ë³´ì„¸ìš”
osaka_urls = check_osaka_coverage()
```

```python
def analyze_klook_categories():
    print("ğŸ” KLOOK ì „ì²´ sitemap ì¹´í…Œê³ ë¦¬ ë¶„ì„...")

    # ë§ˆìŠ¤í„° sitemapì—ì„œ ëª¨ë“  ì¹´í…Œê³ ë¦¬ í™•ì¸
    response = requests.get("https://www.klook.com/ko/sitemap-master-index.xml")
    root = ET.fromstring(response.content)

    sitemap_categories = []
    for sitemap in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
        url = sitemap.text
        category = url.split('sitemap-')[-1].replace('-plain.xml', '')
        sitemap_categories.append(category)

    print("ğŸ“‹ KLOOK ì „ì²´ ì¹´í…Œê³ ë¦¬:")
    for i, category in enumerate(sitemap_categories):
        print(f"  {i+1}. {category}")

    # ì›í•˜ëŠ” ì¹´í…Œê³ ë¦¬ ë§¤í•‘
    target_categories = {
        "experiences-activity": "íˆ¬ì–´/ì–´íŠ¸ë™ì…˜/í…Œë§ˆíŒŒí¬",
        "wifi-sim-card": "ìœ ì‹¬/WiFi",
        "car-rental": "ë ŒíŠ¸ì¹´",
        "mobility": "ì§€í•˜ì² íŒ¨ìŠ¤/êµí†µ"
    }

    excluded_categories = {
        "hotel": "ìˆ™ë°• (ì œì™¸)",
        "flight": "í•­ê³µ (ì œì™¸ - í•˜ì§€ë§Œ sitemapì— ì—†ì„ ê°€ëŠ¥ì„±)"
    }

    print(f"\nğŸ¯ í¬ë¡¤ë§ ëŒ€ìƒ ì¹´í…Œê³ ë¦¬:")
    for category, description in target_categories.items():
        if category in sitemap_categories:
            print(f"  âœ… {category}: {description}")
        else:
            print(f"  âŒ {category}: {description} (sitemapì— ì—†ìŒ)")

    return target_categories

# ì‹¤í–‰í•´ë³´ì„¸ìš”
categories = analyze_klook_categories()
```

```python
def get_osaka_by_categories(target_categories=None):
    """
    ë¸Œë¼ìš°ì € ì‚¬ìš© ì—†ì´ ì˜¤ì‚¬ì¹´ì˜ ì¹´í…Œê³ ë¦¬ë³„ ìƒí’ˆ URL ìˆ˜ì§‘
    """
    print("ğŸš€ ì˜¤ì‚¬ì¹´ ì¹´í…Œê³ ë¦¬ë³„ ìƒí’ˆ URL ìˆ˜ì§‘ (ë¸Œë¼ìš°ì € ì—†ìŒ)...")

    if target_categories is None:
        target_categories = [
            "experiences-activity",  # íˆ¬ì–´/ì–´íŠ¸ë™ì…˜
            "wifi-sim-card",        # ìœ ì‹¬
            "car-rental",           # ë ŒíŠ¸ì¹´
            "mobility"              # êµí†µíŒ¨ìŠ¤
        ]

    osaka_keywords = ['osaka', 'ì˜¤ì‚¬ì¹´', 'japan', 'ì¼ë³¸']
    categorized_urls = {}

    for category in target_categories:
        sitemap_url = f"https://www.klook.com/ko/sitemap-{category}-plain.xml"

        try:
            print(f"  ğŸ“‚ {category} ì¹´í…Œê³ ë¦¬ ì²˜ë¦¬ ì¤‘...")
            response = requests.get(sitemap_url, timeout=30)

            if response.status_code == 200:
                root = ET.fromstring(response.content)
                category_urls = []

                for url_elem in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
                    url = url_elem.text
                    if any(keyword in url.lower() for keyword in osaka_keywords):
                        category_urls.append(url)

                categorized_urls[category] = category_urls
                print(f"    âœ… {len(category_urls)}ê°œ URL ë°œê²¬")

            else:
                print(f"    âŒ {category} sitemap ì ‘ê·¼ ì‹¤íŒ¨ (ìƒíƒœ: {response.status_code})")
                categorized_urls[category] = []

        except Exception as e:
            print(f"    âŒ {category} ì²˜ë¦¬ ì˜¤ë¥˜: {e}")
            categorized_urls[category] = []

    # ê²°ê³¼ ìš”ì•½
    total_urls = sum(len(urls) for urls in categorized_urls.values())
    print(f"\nğŸ“Š ì˜¤ì‚¬ì¹´ ì¹´í…Œê³ ë¦¬ë³„ ìˆ˜ì§‘ ê²°ê³¼:")
    print(f"   ğŸ¯ ì´ {total_urls}ê°œ URL ìˆ˜ì§‘ ì™„ë£Œ")

    for category, urls in categorized_urls.items():
        if urls:
            print(f"   ğŸ“‚ {category}: {len(urls)}ê°œ")
            print(f"      ìƒ˜í”Œ: {urls[0] if urls else 'None'}")

    return categorized_urls

# ì‹¤í–‰í•´ë³´ì„¸ìš”
osaka_categorized = get_osaka_by_categories()
```
