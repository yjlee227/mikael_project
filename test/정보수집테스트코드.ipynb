```python
import requests
import time

# 테스트할 sitemap URLs
sitemap_urls = [
    "https://www.klook.com/ko/sitemap-master-index.xml",
    "https://www.klook.com/sitemap-master-index.xml",
    "https://www.klook.com/robots.txt",
    "https://www.klook.com/ko/robots.txt"
]

for url in sitemap_urls:
    try:
        print(f"\n🔍 테스트: {url}")
        response = requests.get(url, timeout=10, headers={
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        })

        print(f"   상태코드: {response.status_code}")
        if response.status_code == 200:
            content = response.text[:200]
            print(f"   내용: {content}...")

            if 'sitemap' in content.lower() or '<?xml' in content:
                print(f"   ✅ XML/Sitemap 형식 확인됨")
            else:
                print(f"   ❌ XML 형식이 아님")
        else:
            print(f"   ❌ 접근 실패")

    except Exception as e:
        print(f"   ❌ 오류: {e}")

    time.sleep(2)
```

```python
import requests
import xml.etree.ElementTree as ET

# 한국어 sitemap에서 개별 sitemap 목록 가져오기
def get_klook_sitemap_list():
    response = requests.get("https://www.klook.com/ko/sitemap-master-index.xml")
    root = ET.fromstring(response.content)

    sitemap_urls = []
    for sitemap in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
        sitemap_urls.append(sitemap.text)

    print(f"📄 총 {len(sitemap_urls)}개 sitemap 발견:")
    for i, url in enumerate(sitemap_urls[:10]):  # 처음 10개만 표시
        print(f"  {i+1}. {url}")

    return sitemap_urls
sitemap_list = get_klook_sitemap_list()
```

```python
import requests
import xml.etree.ElementTree as ET

def get_seoul_activities():
    print("🎯 서울 액티비티 URL 추출 중...")

    # 액티비티 sitemap에서 서울 관련 URL 찾기
    response = requests.get("https://www.klook.com/ko/sitemap-experiences-activity-plain.xml")
    root = ET.fromstring(response.content)

    seoul_urls = []
    seoul_keywords = ['seoul', '서울', 'korea', '한국']

    for url_elem in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
        url = url_elem.text

        # 서울 관련 URL 필터링
        if any(keyword in url.lower() for keyword in seoul_keywords):
            seoul_urls.append(url)

    print(f"📊 서울 액티비티 URL: {len(seoul_urls)}개 발견")
    for i, url in enumerate(seoul_urls[:10]):  # 처음 10개만 표시
        print(f"  {i+1}. {url}")

    return seoul_urls

# 실행
seoul_activities = get_seoul_activities()
```

```python
import requests
import xml.etree.ElementTree as ET

def check_osaka_coverage():
    print("🔍 오사카 KLOOK 상품 커버리지 분석...")

    # sitemap에서 오사카 관련 URL 수집
    response = requests.get("https://www.klook.com/ko/sitemap-experiences-activity-plain.xml")
    root = ET.fromstring(response.content)

    osaka_keywords = ['osaka', '오사카', 'japan', '일본']
    osaka_urls = []

    for url_elem in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
        url = url_elem.text
        if any(keyword in url.lower() for keyword in osaka_keywords):
            osaka_urls.append(url)

    print(f"📊 오사카 관련 URL: {len(osaka_urls)}개 발견")

    # URL 패턴 분석
    url_patterns = {}
    for url in osaka_urls:
        if '/activity/' in url:
            activity_id = url.split('activity/')[-1].split('-')[0]
            url_patterns[activity_id] = url

    print(f"📋 고유 액티비티: {len(url_patterns)}개")
    print("🔗 샘플 URL:")
    for i, url in enumerate(list(osaka_urls)[:5]):
        print(f"  {i+1}. {url}")

    return osaka_urls

# 실행해보세요
osaka_urls = check_osaka_coverage()
```

```python
def analyze_klook_categories():
    print("🔍 KLOOK 전체 sitemap 카테고리 분석...")

    # 마스터 sitemap에서 모든 카테고리 확인
    response = requests.get("https://www.klook.com/ko/sitemap-master-index.xml")
    root = ET.fromstring(response.content)

    sitemap_categories = []
    for sitemap in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
        url = sitemap.text
        category = url.split('sitemap-')[-1].replace('-plain.xml', '')
        sitemap_categories.append(category)

    print("📋 KLOOK 전체 카테고리:")
    for i, category in enumerate(sitemap_categories):
        print(f"  {i+1}. {category}")

    # 원하는 카테고리 매핑
    target_categories = {
        "experiences-activity": "투어/어트랙션/테마파크",
        "wifi-sim-card": "유심/WiFi",
        "car-rental": "렌트카",
        "mobility": "지하철패스/교통"
    }

    excluded_categories = {
        "hotel": "숙박 (제외)",
        "flight": "항공 (제외 - 하지만 sitemap에 없을 가능성)"
    }

    print(f"\n🎯 크롤링 대상 카테고리:")
    for category, description in target_categories.items():
        if category in sitemap_categories:
            print(f"  ✅ {category}: {description}")
        else:
            print(f"  ❌ {category}: {description} (sitemap에 없음)")

    return target_categories

# 실행해보세요
categories = analyze_klook_categories()
```

```python
def get_osaka_by_categories(target_categories=None):
    """
    브라우저 사용 없이 오사카의 카테고리별 상품 URL 수집
    """
    print("🚀 오사카 카테고리별 상품 URL 수집 (브라우저 없음)...")

    if target_categories is None:
        target_categories = [
            "experiences-activity",  # 투어/어트랙션
            "wifi-sim-card",        # 유심
            "car-rental",           # 렌트카
            "mobility"              # 교통패스
        ]

    osaka_keywords = ['osaka', '오사카', 'japan', '일본']
    categorized_urls = {}

    for category in target_categories:
        sitemap_url = f"https://www.klook.com/ko/sitemap-{category}-plain.xml"

        try:
            print(f"  📂 {category} 카테고리 처리 중...")
            response = requests.get(sitemap_url, timeout=30)

            if response.status_code == 200:
                root = ET.fromstring(response.content)
                category_urls = []

                for url_elem in root.findall('.//{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
                    url = url_elem.text
                    if any(keyword in url.lower() for keyword in osaka_keywords):
                        category_urls.append(url)

                categorized_urls[category] = category_urls
                print(f"    ✅ {len(category_urls)}개 URL 발견")

            else:
                print(f"    ❌ {category} sitemap 접근 실패 (상태: {response.status_code})")
                categorized_urls[category] = []

        except Exception as e:
            print(f"    ❌ {category} 처리 오류: {e}")
            categorized_urls[category] = []

    # 결과 요약
    total_urls = sum(len(urls) for urls in categorized_urls.values())
    print(f"\n📊 오사카 카테고리별 수집 결과:")
    print(f"   🎯 총 {total_urls}개 URL 수집 완료")

    for category, urls in categorized_urls.items():
        if urls:
            print(f"   📂 {category}: {len(urls)}개")
            print(f"      샘플: {urls[0] if urls else 'None'}")

    return categorized_urls

# 실행해보세요
osaka_categorized = get_osaka_by_categories()
```
